## 4. Basic Concepts

### Learning Objectives:
- Understand nodes, edges, and state
- Learn how to construct a simple graph
- Explore the LangGraph API basics

In LangGraph, applications are built as graphs composed of nodes connected by edges. Each node performs a specific operation, and edges define how data flows between nodes.

```python
# Basic example of a LangGraph
from typing import TypedDict, Annotated
from langgraph.graph import StateGraph
from langchain_core.tools import tool

# Define the state structure
class State(TypedDict):
    value: int
    messages: list

# Define a simple node function
def increment_node(state):
    return {"value": state["value"] + 1}

# Create the graph
workflow = StateGraph(State)
workflow.add_node("increment", increment_node)
workflow.set_entry_point("increment")
workflow.set_finish_point("increment")

# Compile the graph
app = workflow.compile()

# Test the graph
result = app.invoke({"value": 0, "messages": []})
print(result)
```
